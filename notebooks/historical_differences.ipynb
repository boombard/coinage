```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
coinfloor = pd.read_csv('../data/coinfloor_historical.csv', index_col=0, parse_dates=True)
bitx = pd.read_csv('../data/bitx.csv', index_col=0, parse_dates=True)
```

```python
compare_df = pd.merge(coinfloor, bitx, how='inner',
                      left_index=True, right_index=True,
                      suffixes=('_coinfloor', '_bitx'))
```

```python
compare_df['avg_difference'] = compare_df['avg_bitx'] - compare_df['avg_coinfloor']
```

```python
compare_df.loc['2016-01-01':, :].plot(y='avg_difference')
```
